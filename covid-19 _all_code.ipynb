```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('country_vaccinations_dataset.csv')
```

```python
df.head()
```

```python
#changing nan values with min numbers of each country
listof = df.columns.difference(['country'])
df1 = df.groupby('country')[listof].transform('min').fillna(0)
df[listof] = df[listof].fillna(df1)
df.head(30)
```

```python
#country does not have any valid vaccination number so code put 0
df.loc[750]
```

```python
#medians for daily vaccination numbers for all countries
groups=df.groupby('country', as_index=False).median()
```

```python
groups.info()
```

```python
groups[['daily_vaccinations']]
```

```python
groups[['daily_vaccinations']].max()
```
